```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.arange(0, 6)
y = np.arange(0, 11, 2)
df = pd.DataFrame({ "x": x, "y": y },)
```

```python
df
```

```python
z = np.linspace(0., 1., 6)
```

```python
plt.scatter(df["x"], df["y"], c=z, cmap="jet")
```

```python
df.plot(kind="scatter", x="x", y="y")
```

```python
df.plot(kind="scatter", x="x", y="y", c=z, colormap="jet")
```
