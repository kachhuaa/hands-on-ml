```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.arange(0, 6)
y = np.arange(0, 11, 2)
z = np.arange(0, 16, 3)
df = pd.DataFrame({ "x": x, "y": y, "z": z })
```

```python
df
```

```python
z = np.linspace(0., 1., 6)
```

```python
plt.scatter(df["x"], df["y"], c=z, cmap="jet")
```

```python
df.plot(kind="scatter", x="x", y="y")
```

```python
df.plot(x="x", cmap=plt.get_cmap("rainbow"))
```
